```python
filepath='/content/car_price_prediction_.csv'
df=pd.read_csv(filepath)
df.head(10)
```

```python
print("Duplicated: ",df.duplicated().sum())
df.shape
```

```python
X = df.drop('Price',axis=1)
Y = df['Price']
print("Shape of X:",X.shape)
print("Shape of Y:",Y.shape)
```

```python
# Identifying categorical columns
cat_cols = X.select_dtypes(include="object").columns
print("Categorical Columns:")
print(cat_cols)
# Performing One-Hot Encoding
X_encoded = pd.get_dummies(X, columns=cat_cols).astype(int)
# Displaying first 5 rows
print(X_encoded.head())
```

```python
from sklearn.preprocessing import StandardScaler
# Numerical columns
num_cols = X_encoded.select_dtypes(include="number").columns
print("Numerical Columns:")
print(num_cols)
# Creating StandardScaler object
scaler = StandardScaler()
# Applying Standard Scaling
X_encoded[num_cols] = scaler.fit_transform(X_encoded[num_cols])
# Showing first 5 rows
print(X_encoded.head())
```

```python
from sklearn.model_selection import train_test_split
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
X_encoded,
Y,
test_size=0.33,
random_state=42
)
# Print the shape of each dataset
print("X_train shape:", X_train.shape)
print("X_test shape :", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape :", y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
# Create the Linear Regression model
model = LinearRegression()
# Train the model
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
# Print first 10 predicted values
print("First 10 Predicted Values:")
print(y_pred[:10])
```

```python
from sklearn.metrics import r2_score
# Calculate R2 Score
r2 = r2_score(y_test, y_pred)
# Print R2 Score
print("R2 Score:", r2)
```

```python
import joblib
# Save model
joblib.dump(model, "LR_ford_car.pkl")
# Save scaler
joblib.dump(scaler, "scaler.pkl")
# Save columns
joblib.dump(X_encoded.columns, "columns.pkl")
```
